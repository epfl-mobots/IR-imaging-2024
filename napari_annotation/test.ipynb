```python
import os
import napari
import cv2
import matplotlib.pyplot as plt

data_path = "../live_bees/h2r4/"
file_you_want = "hive2_rpi4_240419-212901Z.jpg"

ouais = os.path.join(data_path, file_you_want)
img = cv2.imread(ouais, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')

viewer = napari.Viewer()
viewer.add_image(img, name='image1')
viewer.add_layer(napari.layers.Shapes(name='contour'))
```

```python
viewer.layers['contour'].save('contour.csv')
```
